```python
# Creating new dataset with required columns

mldf = pd.DataFrame({   'genre': imt.genre,
                        'duration':imt.duration,
                        'director': imt.director,
                        'production': imt.production_company,
                        'actors':imt.actors,
                        'avgvote':imt.avg_vote,                        
                        })
mldf
```

```python
mldf.isnull().sum()
```

```python
mldf.production.fillna(value= "Unknown", inplace = True)
```

```python
mldf.drop_duplicates(subset=['director','actors'], inplace = True)    # Dropping duplicates 
```

```python
mldf.isnull().sum()
```

```python
mldf
```

```python
# new = data["Name"].str.split(" ", n = 1, expand = True)

actors_new = mldf['actors'].str.split(",", n = 1, expand = True)
actors_new.drop([1], axis = 1, inplace = True)

print(actors_new)

genre_new = mldf['genre'].str.split(",", n = 0, expand = True)
genre_new.drop([1,2], axis = 1, inplace = True)

print(genre_new)
```

```python

mldf_new = pd.DataFrame({   'genre': genre_new[0],
                        'duration':mldf.duration,
                        'director': mldf.director,
                        'production': mldf.production,
                        'actors':actors_new[0],
                        'avgvote':imt.avg_vote,                        
                        })
mldf_new
```

```python
# considering 500 rows as sample

mldf_final = mldf_new[0:500]
mldf_final
```

```python
from sklearn.preprocessing import OneHotEncoder
mldf_encode = pd.get_dummies(mldf_final, columns = ['genre', 'director', 'production', 'actors'])
# df_onehot = pd.get_dummies(df, columns = ['Model', 'Colour'])
mldf_encode.head()
```

```python
mldf_encode.avgvote.describe()
```

```python
mldf_encode1 = mldf_encode.copy()
mldf_encode1
```

```python
bins = [0, 95, 120, 250]
#bin_labels = ['<95', '95-120', '>120']
bin_labels = [0, 1, 2]
mldf_encode1['duration_cat'] = pd.cut(mldf_encode1['duration'], bins=bins, labels=bin_labels, include_lowest=True)

print(mldf_encode1['duration_cat'] )

# Considering 50% percentile as the point to decide whether the movies is successful or not. 
mldf_encode1['avgvote'] = mldf_encode1.avgvote.map(lambda x: 0 if x>6.6 else 1)

mldf_encode1['avgvote'].value_counts()
```

```python
#mldf_encode = mldf_encode.append(mldf_encode1.duration_cat, mldf_encode1.avgvote)

mldf_encode1
```

```python
# mid = df['Mid']
# df.drop(labels=['Mid'], axis=1,inplace = True)
# df.insert(0, 'Mid', mid)
# df

dur = mldf_encode1['duration_cat']
mldf_encode1.drop(labels=['duration'], axis =1, inplace = True)
mldf_encode1.insert(0,'duration', dur)

mldf_encode1
```

```python
# saving the dataframe
mldf_encode1.to_csv('mlecoded.csv')
```

```python

X = mldf_encode1.drop('avgvote',axis='columns')
y = mldf_encode1['avgvote']

print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12) 
```

```python
# Create DecisionTreeclassifer object - With max_depth value
dt_clf = tree.DecisionTreeClassifier(criterion="entropy", splitter = "random", max_depth=9)

# Train the classifier
dt_clf = dt_clf.fit(X_train,y_train) 
```

```python

#Predict the output for test dataset
y_pred = dt_clf.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
abc = metrics.confusion_matrix(y_test, y_pred)
print(abc)
```

```python
disp = metrics.ConfusionMatrixDisplay(abc, display_labels=dt_clf.classes_)
disp.plot()
```

```python
# Alll Scores
metrics.precision_recall_fscore_support(y_test, y_pred)
```
